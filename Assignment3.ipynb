```python
#write python fuction to find the max of three number
def biggest(x,y,z):
  if x>y and x>z:
    print(x,'is biggest number')
  elif y>z and y>x:
    print(y,'is biggest number')
  else:
    print(z,'is biggest number')
```

```python
x=int(input('Enter first number'))
y=int(input('Enter second number'))
z=int(input('Enter third number'))
biggest(x,y,z)
```

```python
#Write a Python function to sum all the numbers in a list.

def sum_list_numbers(numbers):
    total=0
    for x in numbers:
        total+=x
    return total
sample_list=[18,12,30,10,7]
print(f'sample list:{sample_list}')
print(f'expected output:{sum_list_numbers(sample_list)}')
```
