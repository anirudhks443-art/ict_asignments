```python
8. Area of a Triangle (Using Heron‚Äôs Formula)
A land surveyor wants to calculate the area of an irregular triangular plot.
Write a Python program to find the area of a triangle when all three sides are known,
using Heron‚Äôs formula:
ÔøΩ
ÔøΩ =ùëé+ùëè+ùëê
2
Area = ‚àöùë†(ùë† ‚àíùëé)(ùë†‚àíùëè)(ùë†‚àíùëê)
Sample Input:
Enter side a: 5
Enter side b: 6
Enter side c: 7
Sample Output:
Area of the triangle: 14.70
```

```python
import math

# Get the sides of the triangle from the sample input
a = 5
b = 6
c = 7

# Calculate the semi-perimeter
s = (a + b + c) / 2

# Calculate the area using Heron's formula
area = math.sqrt(s * (s - a) * (s - b) * (s - c))

# Print the area, formatted to two decimal places
print(f"Area of the triangle: {area:.2f}")
```

```python
9. Paint Cost Estimator
its total cost.
A painter charges ‚Çπ25 per square meter. Write a program to calculate the total cost to
paint a rectangular wall. The wall has length and height, and there are two windows each
measuring 1.5m √ó 1.2m that should not be painted. Find the total area to be painted and
Sample Input:
Enter wall length (in meters): 12
Enter wall height (in meters): 8
Sample Output:
Total area to be painted: 92.4 sq.m
Total painting cost: ‚Çπ2310.0
```

```python
# Given inputs
wall_length = 12
wall_height = 8
window_length = 1.5
window_height = 1.2
paint_charge_per_sqm = 25

# Calculate wall area
wall_area = wall_length * wall_height

# Calculate area of one window
one_window_area = window_length * window_height

# Calculate area of two windows
two_windows_area = 2 * one_window_area

# Calculate total area to be painted
total_paintable_area = wall_area - two_windows_area

# Calculate total painting cost
total_painting_cost = total_paintable_area * paint_charge_per_sqm

# Print the results
print(f"Total area to be painted: {total_paintable_area:.1f} sq.m")
print(f"Total painting cost: ‚Çπ{total_painting_cost:.1f}")
```

```python
10. Electricity Bill Calculator
An electricity board charges consumers based on unit consumption. The rate per unit is
‚Çπ5.2. Additionally, there is a fixed charge of ‚Çπ100 on every bill. Write a Python program
that takes the number of units consumed and calculates the total bill amount.
Sample Input:
Enter number of units consumed: 250
Sample Output:
Electricity Bill: ‚Çπ1400.0
```

```python
# Given parameters
rate_per_unit = 5.2
fixed_charge = 100

# Get units consumed from the sample input
units_consumed = 250

# Calculate the total bill
total_bill = (units_consumed * rate_per_unit) + fixed_charge

# Print the electricity bill
print(f"Electricity Bill: ‚Çπ{total_bill:.1f}")
```
